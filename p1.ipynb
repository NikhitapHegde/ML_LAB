```python
from queue import PriorityQueue
def best_first(graph,start,goal,heuristic):
    visited=set()
    pq=PriorityQueue()
    pq.put((heuristic[start],start))
    
    while not pq.empty():
        h,node=pq.get()
        if node==goal:
            print ('goal reached')
            return 
        
        if node not in visited:
            for neighbour in graph[node]:
                if neighbour not in visited:
                    pq.put((heuristic[neighbour],neighbour))
                    
            print('visiting node :',node)
            visited.add(node)
            
    print("goal not found")
    return


graph={ 'S':['A','B'],
    'A': ['C', 'D'],
    'B': ['E', 'F'],
    'C': [],
    'D': [],
    'E': ['H'],
    'F': ['I', 'G'],
    'H':[],
    'I':[],
    'G':[],
}

start_node = 'S'
goal_node = 'G'

#Heuristic values from curr node -> goal node
heuristic_values = {
    'S': 13,
    'A': 12,
    'B': 4,
    'C': 7,
    'D': 3,
    'E': 8,
    'F': 2,
    'H': 4,
    'I': 9,
    'G': 0,
}

best_first(graph, start_node, goal_node, heuristic_values)
        
```
